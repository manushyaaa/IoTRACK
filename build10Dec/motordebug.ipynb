```python
#Azimuth Motor Control
for i in range(250 , 234 , -4):
    packet = f'{dt.now().strftime("%H:%M:%S")} {i} 0 ISS_(ZARYA) {MOV_BIT} 0 x\n'
    ser.write(packet.encode('utf-8'))
    print(packet)
    time.sleep(2)
```

```python
import numpy as np

azi1 = np.arange(238, 360, 0.67)
azi2 = np.arange(0,60, 0.67)

azi = np.concatenate((azi1, azi2), axis=None)
ele = np.arange(0, 90, 0.67)

#create a zip of azi and ele 
azi_ele = zip(azi, ele)

#convert to list
azi_ele = list(azi_ele)

for i in azi_ele:
    packet = f'{dt.now().strftime("%H:%M:%S")} {i[0]} {i[1]} ISS_(ZARYA) NEU 0 x\n'
    ser.write(packet.encode('utf-8'))
    print(packet)
    time.sleep(5)
```
